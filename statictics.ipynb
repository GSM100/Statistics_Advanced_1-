1) What is a random variable in probability theory?
-    A random variable is a function that assigns numerical values to the outcomes of a random experiment. It can be discrete (taking specific values) or continuous (taking any value within a range).




2) What are the types of random variables?

-    Discrete Random Variable (takes countable values, e.g., number of heads in a coin toss).

-  Continuous Random Variable (takes any value in a given range, e.g., height of students in a class).

3)  What is the difference between discrete and continuous distributions?

-    Discrete distribution represents probabilities for discrete random variables (e.g., binomial, Poisson).

-  Continuous distribution represents probabilities over an interval (e.g., normal, exponential).

4) What are probability distribution functions (PDF)?
-    A Probability Distribution Function describes the likelihood of a random variable taking specific values. For continuous distributions, the Probability Density Function (PDF) gives the relative likelihood.

5) How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

-   PDF gives the probability of a specific value.

-  CDF gives the probability that a random variable is less than or equal to a given value.

13) What is the Central Limit Theorem (CLT), and why is it critical in statistics?

2. Generate a discrete uniform distribution and plot the PMF

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

x = np.arange(1, 7)  # Example: rolling a fair die (1 to 6)
pmf = stats.randint.pmf(x, 1, 7)  # Discrete uniform PMF

plt.bar(x, pmf, color='blue', alpha=0.6)
plt.xlabel("Values")
plt.ylabel("Probability")
plt.title("PMF of Discrete Uniform Distribution")
plt.show()
```

3. Function to calculate the PDF of a Bernoulli distribution

```python
import scipy.stats as stats

def bernoulli_pdf(p):
    x = [0, 1]  # Bernoulli takes only 0 or 1
    pmf_values = stats.bernoulli.pmf(x, p)
    return dict(zip(x, pmf_values))

print(bernoulli_pdf(0.5))  # Example with p=0.5
```

9. Function to calculate Z-scores from a dataset and plot them

```python
def calculate_z_scores(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 10, 1000)
z_scores = calculate_z_scores(data)

plt.hist(z_scores, bins=30, color='cyan', alpha=0.6)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Z-score Distribution")
plt.show()
```

10. Implement the Central Limit Theorem for a non-normal distribution

```python
samples = 1000
sample_size = 30
population = np.random.exponential(scale=2, size=10000)  # Non-normal population

sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(samples)]

plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='green')
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title("Central Limit Theorem Demonstration")
plt.show()
```

15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem

18. Calculate the Z-score for a data point and compare to standard normal

```python
def z_score(x, mean, std):
    return (x - mean) / std

data = np.random.normal(50, 10, 1000)
z_scores = z_score(data, np.mean(data), np.std(data))

plt.hist(z_scores, bins=30, color='cyan', alpha=0.6)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Z-score Distribution")
plt.show()
```

19. Implement hypothesis testing using Z-statistics

```python
from scipy.stats import norm

sample = np.random.normal(50, 10, 30)
pop_mean = 50
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(len(sample)))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Z-Statistic: {z_stat}, P-Value: {p_value}")
```

20. Create a confidence interval for a dataset

```python
import scipy.stats as stats

sample = np.random.normal(50, 10, 30)
confidence = 0.95
mean, sem = np.mean(sample), stats.sem(sample)
interval = stats.t.interval(confidence, len(sample)-1, loc=mean, scale=sem)

print(f"95% Confidence Interval: {interval}")
```

21. Generate normal data and interpret the confidence interval

25. Compare the standard deviations of two datasets

```python
data1 = np.random.normal(50, 10, 1000)
data2 = np.random.normal(60, 15, 1000)

std1, std2 = np.std(data1), np.std(data2)
print(f"Standard Deviation of Dataset 1: {std1}")
print(f"Standard Deviation of Dataset 2: {std2}")

plt.hist(data1, bins=30, alpha=0.6, label="Dataset 1")
plt.hist(data2, bins=30, alpha=0.6, label="Dataset 2")
plt.legend()
plt.title("Comparison of Standard Deviations")
plt.show()
```

26. Calculate range and interquartile range (IQR)

```python
data = np.random.normal(50, 10, 1000)
range_val = np.max(data) - np.min(data)
iqr = np.percentile(data, 75) - np.percentile(data, 25)

print(f"Range: {range_val}, IQR: {iqr}")
```

27. Implement Z-score normalization and visualize transformation

```python
def z_score_normalization(data):
    return (data - np.mean(data)) / np.std(data)

data = np.random.normal(50, 10, 1000)
normalized_data = z_score_normalization(data)

plt.hist(normalized_data, bins=30, color='blue', alpha=0.6)
plt.xlabel("Z-score Normalized Values")
plt.ylabel("Frequency")
plt.title("Z-score Normalization")
plt.show()
```

28. Calculate skewness and kurtosis of a normal dataset
